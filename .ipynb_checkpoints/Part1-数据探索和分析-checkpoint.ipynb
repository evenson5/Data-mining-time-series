#### 一 、赛题简介
蚂蚁金服的上亿会员涉及大量的资金流入和流出，庞大用户群对应资金管理的巨大压力。在即保证资金流动性风险最小，又满足日常业务运转，精准地预测资金流入流出最重要。

大赛希望参赛者通过对例如余额宝用户的申购和赎回数据的把握，精准预测未来每日的资金流入流出。（货币基金：资金流入=申购行为，资金流出=赎回行为）。

**重点**
- 目的：精准地预测资金流入流出额度。
- 余额宝的申购赎回金额的精准把握。
- 资金流入：申购；资金流出：赎回。
- 利用2013年7月-2014年8月的数据，**预测2014年9月每一天的申购和赎回的总量**

- 个人形式参赛，对象不限。

### 1.2 指标分析
整个赛题初步理解后，要对赛题的最终评判即指标进行分析。本体的评判指标如下：

评估指标的设计主要期望选手*对未来 30 天内***每一天申购和赎回的总量数据预测的越准越好**，同时考虑到可能存在的*多种情况*。

譬如有些选手在 30 天中 29 天预测都是非常精准的但是某一天预测的结果可能误差很大，而有些选手在 30 天中每天的预测都不是很精准误差较大，如果采用绝对误差则可能导致前者的成绩比后者差，而在实际业务中可能更倾向于前者。所以最终选用**积分式**的计算方法：

*每天的误差选用相对误差来计算，然后根据用户预测申购和赎回的相对误差，通过得分函数映射得到一个每天预测结果的得分，将 30 天内的得分汇总，然后结合实际业务的倾向，对申购赎回总量预测的得分情况进行加权求和，得到最终评分*。


具体操作如下：
1) 计算所有用户在测试集上每天的申购及赎回总额与实际情况总额的误差。
每日申购相对误差（真实值$Z_{i}$，预测值$\hat{Z_{i}}$）:
                $ Purchase_{i} = \frac{\left |Z_{i}-\hat{Z_{i}} \right |}{Z_{i}}$
                
每日赎回相对误差（真实值$y_{i}$，预测值$\hat{y_{i}}$）：
                $ Redeem_{i} = \frac{\left |y_{i}-\hat{y_{i}} \right |}{y_{i}}$

2) 申购预测得分与 Purchasei 相关，赎回预测得分与 Redeemi 相关 , 误差与得分之间的计算公式不公布，*但保证该计算公式为单调递减的，即误差越小，得分越高，误差越大，得分越低。*当第 i 天的申购误差 Purchasei =0 ，这一天的得分为 10 分；当 Purchasei > 0.3 ，其得分为 0 。

3) 最后公布总积分 = 申购预测得分 *45%+ 赎回预测得分 *55% 。 

#### 上述总结
1. 每一天的预测误差都较小
2. 积分制：每天最高10分，最低0分，共300分
3. 误差较小，当评估指标小于0.1，则为10分；误差较大当评估指标大于0.3则为0分。
4. 两个公式，一个是每天的相对误差，分申购和赎回；另一个是总结分计算公式。

### 1.2.1 常见评估指标：
- 评估指标

- 常见分类和回归的评价指标形式：

 *分类算法常见评估指标*
 - 二分类器/分类算法，评价指标主要有`accuracy`,`[Precssion, Recall, F-score,Pr曲线]`，`ROC-AUC曲线`
 - 多分类器/分类算法，评价指标主要有`accuracy`,`[宏平均和微平均，F-score]`
 
 *回归算法常见评估指标*
 - `平均绝对误差（MAE）`，`均方误差（MSE）`,`平均绝对百分误差（MAPE）`，`均方根误差（RMSE）`，`R2（R-square）`

 - 最后的评判指标（积分）权重占比分别为，申购45%，赎回55%，权重有侧重。

### 1.2.2 评估指标的含义

#### 分类指标含义

- TP（True Positive）：预测结果和真实结果均为正例
- TN（True Negative）：预测结果和真实结果均为反例
- FP（False Positive）：把假例预测为正例
- FN（False Negative）：把正例预测为假例

**精准率/查准率：**被分类器判定正例中正样本的比重

$precision_{k} = \frac{TP}{TP + FP}$

**召回率/查全率：**被预测为正例占总的正例的比重

$precision_{k} = \frac{TP}{TP + FN}$

**准确率：**分类器对整个样本判断正确的比重

$accuracy = \frac{TP+TN}{TP + TN + FP + FN}$

**f1-score**

$f1_{k}=\frac{2*precision_{k}*recall_{k}}{precision_{k} + recall_{k}}$

$score=({\frac{1}{n}\sum f1_{k}})^2$

#### 回归指标含义
**均方误差MSE：**应用于线性回归损失函数，易理解。

$MSE = \frac{1}{m}\sum_{i=1}^{m}(y_{i}-\hat{y_{i}})^2$

**均方根误差RMSE：**与MSE实质一样，更易于数据的描述，且平滑可微。

$RMSE = \sqrt{\frac{1}{m}\sum_{i=1}^{m}(y_{i}-\hat{y_{i}})^2}$

**平均绝对误差MAE：**预测值与真实值之间绝对误差的均值，惩罚度低于前两者。

$MAE = \frac{1}{m}\sum_{i=1}^{m} \left |y_{i}-\hat{y_{i}}\right |$

**平均绝对百分误差MAE：**应用于每次与真实值的距离计算，相比适用于本题。

$MAPE = \sum_{i=1}^{m} \left |y_{i}-\hat{y_{i}}\right |\frac{100}{m}$

**R-square:**线性回归的评估标准。

SSE（残差平方和）

SSR（回归平方和）

SST（总平方和）

$SSE =  \sum_{i=1}^{n} \omega _{i}(y_{i}-\hat{y_{i}})$

$SSR =  \sum_{i=1}^{n} \omega _{i}(\hat{y_{i}}-\bar{y_{i}})$

$SST =  \sum_{i=1}^{n} \omega _{i}(y_{i}-\bar{y_{i}})^2$

$R^2 = \frac{SSR}{SST} = \frac{SST-SSE}{SST}=1 - \frac{SSE}{SST}$

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

#### 分类指标平价计算示例

```python
###二分类/多分类指标：accuracy
y_pred = [0,1,0,1]
y_ture = [0,1,1,1]
print('ACC:',accuracy_score(y_ture,y_pred))
```

```python
### 二分类/多分类指标：F1-score
from sklearn import metrics
y_pred = [0,1,0,0]
y_true = [0,1,0,1]
print('Precision',metrics.precision_score(y_true,y_pred))
print('Recall',metrics.recall_score(y_true,y_pred))
print('F1-score:',metrics.f1_score(y_true,y_pred))
```

```python
### 二分类AUC
from sklearn.metrics import roc_auc_score
y_true = np.array([0,0,1,1])
y_scores = np.array([0.1,0.4,0.35,0.8])
print('AUC score:',roc_auc_score(y_true,y_scores))
```

#### 回归指标评价示例

```python
from sklearn import metrics

### MAPE实现
def MAPE(y_true, y_pred):
    return np.mean(np.abs((y_pred - y_true) / y_true))

y_true = np.array([1.0,5.0,4.0,3.0,2.0,5.0,-3.0])
y_pred = np.array([1.0,4.8,3.6,3.2,3.0,4.6,-2.0])

## MSE
print('MSE:',metrics.mean_squared_error(y_true, y_pred))
## RMSE
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_true, y_pred)))
## MAE
print('MAE:',metrics.mean_absolute_error(y_true, y_pred))
##MAPE
print('MAPE:',MAPE(y_true, y_pred))
```

```python
### R2-score
from sklearn.metrics import r2_score
y_true = [3,-0.5,2,7]
y_pred = [2.5,0.0,2,8.0]
print('R2_score:',r2_score(y_true, y_pred))
```

```python
# 使用OneHot方法将翌日特征划分，获取划分后特征

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
total_balance = total_balance.reset_index()
week_feature = encoder.fit_transform(np.array(total_balance['weekday']).reshape(-1, 1)).toarray()
week_feature = pd.DataFrame(week_feature,columns=['weekday_onehot']*len(week_feature[0]))
feature = pd.concat([total_balance, week_feature], axis = 1)[['今日总购买量=直接购买+收益', '今日总赎回量=消费+转出','weekday_onehot','date']]
feature.columns = list(feature.columns[0:2]) + [x+str(i) for i,x in enumerate(feature.columns[2:-1])] + ['date']
```
